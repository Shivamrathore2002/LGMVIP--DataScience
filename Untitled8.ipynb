```python
#importing libraries
import numpy as np
import imageio
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/20210424_184745.jpg")
```

```python
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_image)
```

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image)
```

```python
blured = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
inverted_blur = 255 - blured
sketch = cv2.divide(gray_image, inverted_blur, scale = 256.0)
```

```python
plt.imshow(sketch)
```

```python
def dodgeV2(X,Y):
  return cv2.divide(X, 255- Y , scale =256)
```

```python
pencil = dodgeV2(gray_image, blured)
plt.imshow(pencil)
```

```python
pencil= dodgeV2(gray_image, blured)
plt.imshow(pencil)
```
